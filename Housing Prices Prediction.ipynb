```python
year = []
for i in range(len(Raw_Housing_Data["Date House was Sold"])):
    k = Raw_Housing_Data["Date House was Sold"][i].split()[-1]
    year.append(k)
```

```python
Raw_Housing_Data["year_sold"] = year
Raw_Housing_Data["year_sold"].head()
```

```python
Raw_Housing_Data["luxury_home"] = 0
for i in range(len(Raw_Housing_Data)):
    count = 0
    if Raw_Housing_Data["Waterfront View"][i] == "Yes":
        count += 1
    if Raw_Housing_Data["Condition of the House"][i] in ["Good", "Excellent"]:
        count += 1
    if Raw_Housing_Data["Overall Grade"][i] >= 8:
        count += 1
    if count >= 2:
        Raw_Housing_Data['luxury_home'][i] = "Yes"
    else:
        Raw_Housing_Data["luxury_home"][i] = "No"
Raw_Housing_Data["luxury_home"].unique()
```

```python
def luxury_home(row):
    count = 0
    if row[0] == "Yes":
        count += 1
    if row[1] in ["Good", "Excellent"]:
        count += 1
    if row[2] >= 8:
        count += 1
    if count >= 2:
        return "Yes"
    else:
        return "No"
Raw_Housing_Data["luxury_home"] = Raw_Housing_Data[["Waterfront View", "Condition of the House", "Overall Grade"]].apply(luxury_home, axis = 1)
Raw_Housing_Data["luxury_home"].unique()
```
